```python
import csv
import json
import glob
import os

# 如果你的 CSV 文件都在同一个目录下，用 glob 匹配
csv_files = ["/mnt/afs/languilin/deer-flow/train.csv","/mnt/afs/languilin/deer-flow/test.csv","/mnt/afs/languilin/deer-flow/validation.csv"]

with open("/mnt/afs/languilin/deer-flow/input_files/travel_plan.jsonl", "w", encoding="utf-8") as jsonlfile:
    for file_path in csv_files:
        with open(file_path, "r", encoding="utf-8") as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                query = row["query"]
                json_line = json.dumps({"query": query}, ensure_ascii=False)
                jsonlfile.write(json_line + "\n")
```
